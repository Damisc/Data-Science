```python
# Importng the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Loading iris dataset

iris = load_iris(as_frame=True)
df = iris.frame
```

```python
target_names = dict(enumerate(iris.target_names))
```

```python
print(target_names)
```

```python
df["species"] = df["target"].map(target_names)
```

```python
# Quick pandas inspection
print("\n Shape:", df.shape)
print("-"*50)
print("\n Info:")
print(df.info())
print("-"*50)
print("\n Describe (numeric):")
print(df.describe())
```

```python
# convert to numpy arrays
feature_cols = iris.feature_names

x = df[feature_cols]
y = df["target"]
```

```python
print(x)
```

```python
print(y)
```

```python
X = x.to_numpy()
Y = y.to_numpy()
```

```python
# Numpy stats {column wise}: mean, std

col_mean = X.mean(axis=0)
col_std = X.std(axis=0, ddof=0)

print("\n Numpy column mean:", col_mean)
print("\n Numpy column std:", col_std)
```

```python
# Missing values in the dataframe

print("Missing values in the dataframe:")
print(df.isnull().sum())
```

```python
df.groupby("species")[feature_cols].mean()
```

```python
# Pair plot with seaborn

sns.set_style("whitegrid")

sns.pairplot(df[feature_cols + ["species"]], hue="species", diag_kind="kde")
plt.suptitle("Pairwise feature Relationships", y=1.02)
plt.show()
```

```python
# Correlation heatmap

corr = df[feature_cols].corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation - Heatmap")
plt.show()
```

```python
# Count plao

plt.figure(figsize=(6,4))
sns.countplot(x="species", data=df)
plt.title("Class Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Scater plot

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="petal length (cm)", y="petal width (cm)", hue="species")
plt.title("Petal Length vs Petal width by Species")
plt.tight_layout()
plt.show()
```
